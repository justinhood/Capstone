```python
P=np.empty([10,10])
#63.12
for i in range(0,10):
    for j in range(0,10):
        P[i,j]=55.99*total[i,j]-748.92*(i+1)-161.58*(j+1)
```

```python
ay=sns.heatmap(P,cmap="gist_heat",linewidths=.5,center=0)
ay.set_xlabel("Technicians");
ay.set_ylabel("Pharmacists");
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
Tech=np.arange(1,11)
Pharm=np.arange(1,11)
Tech,Pharm=np.meshgrid(Tech,Pharm)
surf=ax.plot_surface(Tech,Pharm,P, cmap=cm.coolwarm,linewidth=0,antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)
#ax.set_xlim(10,1)
ax.view_init(azim=175)
ax.set_xlabel("Technicians");
ax.set_ylabel("Pharmacists");
```
